```python
import random

suits = ['Clubs', 'Spades', 'Diamonds', 'Hearts']
rank_list = ['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace']
rank_value = {
    'Two' : 2,
    'Three' : 3,
    'Four' : 4,
    'Five' : 5,
    'Six' : 6,
    'Seven' : 7,
    'Eight' : 8,
    'Nine' : 9,
    'Ten' : 10,
    'Jack' : 11,
    'Queen' : 12,
    'King' : 13,
    'Ace' : 14
}

class card:
    def __init__(self, rank, suite):
        self.rank = rank
        self.suite = suite
    
    def __str__(self):
        return(f"Card is {self.rank} of {self.suite}")

class deck:
    def __init__(self,deck_id):
        self.new_deck = []
        self.deck_id = deck_id
        for i in suits:
            for j in rank_list:
                self.new_deck.append((i,rank_value[j]))
    def __str__(self):
        return(str(len(self.new_deck)))
    def shuffle(self):
        random.shuffle(self.new_deck)
        return
    def splitCards(userdeck):
        player1 = userdeck.new_deck[0:26]
        player2 = userdeck.new_deck[27:52]
        return(player1,player2)

class player():
    def __init__(self,Name):
        self.Name = Name
        self.mycards = []
    def deal(cardslist):
        return(cardslist.pop(0))
    def add_card(self,card_now):
        self.mycard.append(card_now)
        return
```

```python
if __name__ == '__main__':
    current_deck = deck('1')
    player1 = player(str(input('Enter Player 1 name')))
    player2 = player('Computer')
    print(player1.Name)
    print(player2.Name)
    player1.mycards = deck.splitCards(current_deck)[0]
    player2.mycards = deck.splitCards(current_deck)[0]

    game_on_flag = 0
    turn = 1
    #initial card
    player2_deal = player.deal(player1.mycards)
    while game_on_flag == 0: 
        if turn == 1:
            print(F"{player1.Name}'s turn")
            try:
                player1_deal = player.deal(player1.mycards)
            except:
                print(f'{player1.Name} is out of cards, Computer wins!')
                game_on_flag = 1
                break
            print(F'Player 1 card : {player1_deal} | Player 2 card : {player2_deal}')
            if player1_deal[1] == player2_deal[1]:
                player1.add_card(player1_deal)
                player1.add_card(player2_deal)
                print('Player 1 Win the deal!')
                turn = 0
            else:
                turn = 0
                continue
        if turn == 0:
            print("Computers's turn")
            try:
                player2_deal = player.deal(player1.mycards)
            except:
                print(f'Computer is out of cards, {player1.Name} wins!')
                game_on_flag = 1
                break
            print(F'Player 1 card : {player1_deal} | Player 2 card : {player2_deal}')
            if player1_deal[1] == player2_deal[1]:
                player2.add_card(player1_deal)
                player2.add_card(player2_deal)
                print('Player 2 Win the deal!')
                turn = 1
            else:
                turn = 1
                continue
        if len(player1.mycard) == 0:
            print('Computer Won!')
            game_on_flag = 1
        elif len(player2.mycard) == 0:
            print(f'{player1.name} won !!')
            game_on_flag = 1
        else:
            continue
```
